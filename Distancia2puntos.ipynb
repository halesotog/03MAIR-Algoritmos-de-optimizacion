### PRIMER INTENTO FUERZA BRUTA

```python
proximo1 = max(LIASTA_1D)+1
proximo2 = 0

for i in LIASTA_1D:
    for j in LIASTA_1D:
        if i != j:
            if abs(i-j) < abs(proximo1 - proximo2):
                proximo1 = i
                proximo2 = j
            
print('Elementos mas proximos1: ' + str(proximo1) +' proximo2: ' + str(proximo2) + ' distancia ' + str(abs(proximo1-proximo2)))
```
